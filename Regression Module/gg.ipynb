```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = np.array([1, 2])
y = np.array([300, 500])
a, b = np.polyfit(X, y, 1)
plt.scatter(X, y)
x1 = min(X)
x2 = max(X)
plt.plot([x1, x2], [a * x1 + b, a * x2 + b], color="red")
plt.show()
```
